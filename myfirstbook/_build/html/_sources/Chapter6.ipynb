### **2. Concept 1: Binary Trees**

#### **2.1. Intuition**

Binary trees are a fundamental data structure in computer science, providing a simple yet powerful way to represent hierarchical relationships. They are widely used in various applications, including image processing, segmentation, information retrieval, and searching algorithms (Salembier & Garrido, 2000; , Senhadji-Navarro & Garcia-Vargas, 2018). Binary trees have also found applications in fields such as mathematics, combinatorics, and phylogenetic reconstruction (Li et al., 2019; , Broutin & Devroye, 2006). The structure of binary trees allows for efficient representation and manipulation of data, making them essential in algorithm design and optimization (Adler & Heeringa, 2011; , Magner et al., 2018).

A binary tree is a discrete structure that consists of nodes, where each node contains a data element and references to up to two children, designated as the left child and the right child. This structure is a key element in various domains of computer science, providing an efficient means of organizing data hierarchically.

**Definition and Properties**
1. Node: The basic unit of a binary tree. Each node comprises:

* Data: An element of any data type (e.g., integer, string) that the node stores.
* Left Child: A reference to the left child node.
* Right Child: A reference to the right child node.
2. Root Node: The initial node from which all other nodes are accessible. The root node does not have a parent.

3. Leaf Nodes: Nodes that do not have any children. They are the endpoints of the tree.

4. Depth of a Node: The number of edges from the node to the tree's root node.

5. Height of a Node: The number of edges on the longest downward path between that node and a leaf. The height of the tree itself is the height of the root.

**Types of Binary Trees**
* Full Binary Tree: Every node has either 0 or 2 children, but never one child alone.
* Complete Binary Tree: All levels are completely filled except possibly the last, which is filled from the left.
* Perfect Binary Tree: A complete binary tree in which all leaf nodes are at the same depth, and in which every parent has exactly two children.
* Balanced Binary Tree: A binary tree is balanced if the height of the left and right subtrees of any node differ by no more than one. This balance is crucial for maintaining efficient operations.
* Binary Search Tree (BST): A binary tree where for each node, the left subtree's values are less than the node's value, and the right subtree's values are greater than the node's value. BSTs support efficient data search, insertion, and deletion operations.

**Applications of Binary Trees**

Binary trees are versatile structures used in numerous applications:
* Sorting and Searching: Binary search trees enable efficient searching akin to binary search in arrays, making them suitable for database indexing.
* Priority Queues: Binary trees are used in the implementation of priority queues, where operations are based on the priority of elements.
* Syntax Trees: Used in compilers to represent the syntactic structure of programming languages.
* Decision Processes: In machine learning, decision trees are used for classification and regression tasks.

**Algorithmic Considerations**

Operations such as insertion, deletion, and traversal (in-order, pre-order, post-order) in binary trees vary in complexity, generally depending on the tree's height and structure. Ensuring efficient performance often involves maintaining tree balance through algorithms such as AVL or Red-Black trees.

In the realm of image processing, binary partition trees have been leveraged as an efficient representation for tasks like image segmentation and information retrieval (Salembier & Foucher, 2016). These trees offer a structured approach to organizing regions within an image, facilitating the extraction of meaningful information from complex visual data (Salembier & Foucher, 2016). Furthermore, in the context of hyperspectral imaging, binary partition trees have been instrumental in image representation and processing, showcasing their versatility across different domains (Veganzones et al., 2014; , Valero et al., 2013).

From a mathematical perspective, binary trees have been extensively studied, with research focusing on limit laws, random binary search trees, and functions of subtrees (Devroye, 1991; , Holmgren & Janson, 2015; , Devroye, 2002). These studies delve into the statistical properties and behaviors of binary trees, shedding light on their probabilistic nature and providing insights into their structural characteristics. Moreover, the use of binary trees extends beyond computer science and mathematics, with applications in gene duplication analysis, classification algorithms, and compression techniques (Chen et al., 2000; , Jin & Mo, 2013; , Radha et al., 1996).

Overall, binary trees serve as a fundamental building block in various disciplines, offering a versatile and efficient means of organizing and processing data. Their simplicity, coupled with their wide-ranging applications, underscores their significance in both theoretical research and practical implementations.

#### **2.2. Example**

Consider organizing a set of numbers: 7, 5, 12, 3, 6, 9, 15 into a binary tree starting from the root (7) and inserting each subsequent number following the binary tree logic (smaller to the left, larger to the right).

#### **2.3. Code**

Here is the Python code to create and traverse a simple binary tree:

```python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key
```

* Node class: Represents a node in a binary tree.

* __init__ method: Initializes a node with a given value (key) and sets both left and right children to None.

```python
def insert(root, key):
    if root is None:
        return Node(key)
    else:
        if root.val < key:
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root
```

* Purpose: Inserts a new node with the specified key into the binary tree rooted at root.
* Recursive nature: The function recursively finds the correct position for the new key, maintaining the binary tree's property (smaller keys on the left, larger keys on the right).

```python
def inorder_traversal(root):
    return inorder_traversal(root.left) + [root.val] + inorder_traversal(root.right) if root else []

# Create root
root = Node(7)
insert(root, 5)
insert(root, 12)
insert(root, 3)
insert(root, 6)
insert(root, 9)
insert(root, 15)

# Traverse tree
print("Inorder Traversal:", inorder_traversal(root))
```

* Purpose: Performs an inorder traversal of the binary tree (visit left child, current node, then right child).
* Recursive structure: Combines the values in a list, which results in displaying all node values in ascending order.

Visualization

```python
import matplotlib.pyplot as plt
import networkx as nx
```

* matplotlib.pyplot is used for plotting graphs visually.
* networkx is employed to manage and visualize complex graphs. In this case, it's used to represent the binary tree structure.

```python
def add_edges(graph, node, pos, x=0, y=0, layer=1):
    pos[node.val] = (x, y)
    if node.left is not None:
        graph.add_edge(node.val, node.left.val)
        l = x - 1 / layer
        add_edges(graph, node.left, pos, x=l, y=y-1, layer=layer+1)
    if node.right is not None:
        graph.add_edge(node.val, node.right.val)
        r = x + 1 / layer
        add_edges(graph, node.right, pos, x=r, y=y-1, layer=layer+1)
```

* graph is a DiGraph object from networkx, representing the tree.
* node refers to the current node in the recursive traversal.
* pos is a dictionary storing positions for each node for plotting. Each node's value is used as a key.
* x and y are coordinates. x is adjusted based on the tree's depth (layer), making sure children don't overlap visually.
* Recursive Calls: The function calls itself for the left and right children of the current node, positioning them increasingly further from the parent as the depth increases, which helps in avoiding overlaps and maintaining a clear structure.

```python
def draw_binary_tree(root):
    graph = nx.DiGraph()
    pos = {}
    add_edges(graph, root, pos)
    nx.draw(graph, pos, with_labels=True, arrows=False, node_size=500, node_color="skyblue")
    plt.title('Binary Tree Visualization')
    plt.show()
    # Assuming the rest of your code is already defined
draw_binary_tree(root)
```

* graph: A new directed graph is created where each node points to its children, reflecting the structure of a binary tree.
* pos: An empty dictionary that add_edges will fill with positions for each node based on their hierarchical relationship.
* nx.draw: Draws the graph using the positions calculated. Nodes are labeled with their values, and the styling makes them easy to distinguish.
* Display: The plt.show() function renders the window showing the tree.

The overall approach uses a mix of tree traversal to construct the graph structure and a recursive positioning algorithm to ensure that the visual representation respects the hierarchical nature of the binary tree. This method is particularly effective for educational purposes, as it clearly demonstrates the structure and relationships within the tree.

Graphs are fundamental data structures in computer science, used extensively to model relationships and solve complex problems across various domains such as social networks, transportation, and internet infrastructure.

#### **3.2. Example**

Imagine a small network of friends where connections represent friendships. This can be represented as a graph where each person is a node and each friendship is an edge.

#### **3.3. Code**

Here's how you could implement a simple graph in Python and perform a breadth-first search (BFS):

```python
from collections import deque

class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v):
        if u in self.graph:
            self.graph[u].append(v)
        else:
            self.graph[u] = [v]
```

* Graph class: Represents a simple directed graph using an adjacency list.
* __init__ method: Initializes an empty dictionary graph to store graph vertices and edges.
* add_edge method: Adds a directed edge from vertex u to vertex v. If u is not already in graph, it initializes a new list.

```python
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            visited.add(vertex)
            queue.extend(set(graph.graph.get(vertex, [])) - visited)
    return list(visited)

# Create graph
g = Graph()
g.add_edge('Alice', 'Bob')
g.add_edge('Alice', 'Cindy')
g.add_edge('Bob', 'Dan')
g.add_edge('Dan', 'Alice')

# Perform BFS
print("BFS starting from Alice:", bfs(g, 'Alice'))
```

* Purpose: Performs a breadth-first search starting from the start vertex.
* deque for queuing: Ensures that vertices are visited in the order they are discovered.
* Handling visited nodes: Prevents revisiting nodes, crucial for avoiding infinite loops in cyclic graphs.
* Efficient lookup and extension: Uses sets to quickly check and add vertices that have not been visited yet.

**Visualization**

```python
import matplotlib.pyplot as plt
import networkx as nx
def draw_graph(graph, start_vertex, bfs_order):
    G = nx.DiGraph()  # Create a directed graph
    for node in graph.graph:
        for neighbor in graph.graph[node]:
            G.add_edge(node, neighbor)  # Add edges to the graph

    pos = nx.spring_layout(G)  # Positions for all nodes

    # Draw the network graph
    nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', edge_color='#3b0a45', width=2, arrowstyle='-|>', arrowsize=10)

    # Highlight the BFS traversal path
    path_edges = list(zip(bfs_order[:-1], bfs_order[1:]))
    nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=2, style='dashed')

    plt.title('Graph Visualization with BFS Traversal')
    plt.show()
# Assuming the rest of your code is already defined
bfs_result = bfs(g, 'Alice')
draw_graph(g, 'Alice', bfs_result)
```

* Graph Creation: The function creates a DiGraph object from networkx to represent the directed nature of your graph.
* Adding Edges: Loops through your custom graph structure and adds all edges to the networkx graph.
* Positioning: Uses spring_layout for aesthetically pleasing automatic positioning of nodes.
* Drawing the Graph: Draws the graph with specific styling for nodes and edges.
* Highlighting BFS Path: Converts the BFS traversal order into a list of edges and redraws these edges with distinct styling (red and dashed) to emphasize the BFS path.

#### **Problem 1: Maximum Depth of Binary Tree**

**4.1.1. Problem Statement**

Find the maximum depth of a binary tree.

**4.1.2. Intuition**

The depth of a binary tree is the longest path from the root node down to the farthest leaf node.

**4.1.3. Code**

```python
def maxDepth(root):
    if not root:
        return 0
    else:
        left_height = maxDepth(root.left)
        right_height = maxDepth(root.right)
        return 1 + max(left_height, right_height)

# Example tree (constructed previously)
print("Maximum Depth:", maxDepth(root))
```

* Base Case: If the root is None (i.e., the tree is empty or it's an empty leaf), return 0.
* Recursive Calls: Calculate the depth of the left subtree (left_height) and the right subtree (right_height) by recursively calling maxDepth.
* Compute Maximum Depth: Add 1 to the maximum of left_height and right_height. The addition of 1 represents the current node in the path length calculation.

This function utilizes a depth-first search approach, specifically a post-order traversal, to reach the leaf nodes and then bubbles up the maximum depth values to the root.

#### **Problem 2: Find if Path Exists in Graph**

**4.2.1. Problem Statement**

Given a directed graph, determine if a path exists between two nodes.

**4.2.2. Intuition**

Use BFS or DFS to explore the graph from the starting node and check if the destination node can be reached.

**4.2.3. Code**

```python
def hasPath(graph, start, end):
    visited = set()
    stack = [start]
    while stack:
        node = stack.pop()
        if node == end:
            return True
        if node not in visited:
            visited.add(node)
            stack.extend(graph.graph.get(node, []))
    return False

# Using previously created graph
print("Path exists from Dan to Alice:", hasPath(g, 'Alice', 'Cindy'))
```

* Stack Initialization: Start with a stack containing the start vertex.
* Loop Through Stack: Continue until the stack is empty, indicating all reachable nodes have been processed.
* Check End Condition: If the current node is the end vertex, return True indicating a path exists.
* Mark as Visited and Explore Further: If the node hasn't been visited, mark it as visited and add its neighbors to the stack for further exploration.
  
This code implements a depth-first search (DFS) to find a path in the graph. The use of a stack ensures nodes are explored in a LIFO (last-in-first-out) order.

```python
def draw_graph_with_path(graph, start, end, path_exists):
    G = nx.DiGraph()
    for node in graph.graph:
        for neighbor in graph.graph[node]:
            G.add_edge(node, neighbor)

    pos = nx.spring_layout(G)  # Positions for all nodes using the spring layout
    nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', edge_color='gray', width=2, arrowstyle='-|>', arrowsize=10)

    if path_exists:
        # To highlight the path, you need to extract the path from 'start' to 'end'
        # This requires a modification to 'hasPath' to store the path or using nx.shortest_path if path exists
        try:
            path = nx.shortest_path(G, start, end)
            path_edges = list(zip(path, path[1:]))
            nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=2, style='dashed')
        except nx.NetworkXNoPath:
            print("No path exists.")

    plt.title('Graph Visualization with Path Highlight')
    plt.show()
path_exists = hasPath(g, 'Alice', 'Dan')
draw_graph_with_path(g, 'Alice', 'Dan', path_exists)
```

#### **Problem 3: Cycle Detection in Directed Graph**

**4.3.1. Problem Statement**

Detect if a directed graph contains a cycle.

**4.3.2. Intuition**

A cycle in a graph means revisiting a node. Use DFS to track visited nodes and backtrack to identify cycles.

**4.3.3. Code**

```python
def detectCycle(graph):
    def dfs(node, visited, recStack):
        visited.add(node)
        recStack.add(node)
        for neighbour in graph.graph.get(node, []):
            if neighbour not in visited:
                if dfs(neighbour, visited, recStack):
                    return True
            elif neighbour in recStack:
                return True
        recStack.remove(node)
        return False

    visited, recStack = set(), set()
    for node in list(graph.graph):
        if node not in visited:
            if dfs(node, visited, recStack):
                return True
    return False

# Detect cycle in the graph
print("Cycle detected:", detectCycle(g))
```

* Recursive DFS Function: dfs is defined to explore each node. It uses two sets, visited to track all visited nodes, and recStack to track nodes currently in the recursive call stack.
* Cycle Detection Logic:
* * If a node in recStack is encountered again, it indicates a back edge, which is a cycle in a directed graph.
* * Nodes are added to recStack when they are visited and removed when the recursion for that node returns.
* Global Loop: Iterates over each node to handle disconnected components and ensure all nodes are checked for cycles.
The detection of cycles in this manner is crucial for understanding dependencies and prerequisites in systems like task scheduling, course planning, and more, where a cyclic dependency can lead to deadlocks or inconsistencies. This DFS-based approach efficiently identifies cycles by tracing the recursion path and checking for revisited nodes within the same traversal path.
