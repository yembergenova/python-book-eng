# Brief History of Python programming language.

Python is a popular high-level, general-purpose programming language. In 1991, Guido van Rossum created the original concept, which the Python Software Foundation later improved. Programmers may express concepts in fewer lines of code thanks to its syntax, which was primarily built with code readability in mind.

History was about to be written in the late 1980s. That was the beginning of the Python project. Not long afterward, in December 1989, Guido Van Rossum started working on application-based projects at Centrum Wiskunde & Informatica (CWI), a Dutch organization. He began it as a hobby project since he was trying to find something engaging to do over Christmas.

1991 saw the official release of the language. When compared to Java, C++, and C, it utilized a lot less code to describe the concepts at the time of release. Its design ethos was also excellent. Code readability and increased developer productivity are its primary goals. When it was first launched, it was more than capable of supporting various fundamental data types for exception handling, inheritance in classes, and other features. 


# Main Key Features of Python programming language.

**Free and Open Source.** Python language is freely available at the official website and you can download it from the given download link, click on the Download Python keyword Download Python. Since it's open-source, the public can access the source code as well. Thus, you are able to download, use, and distribute it. 

**Simple to write code for.** Compared to other programming languages like C, C#, Javascript, Java, etc., Python is a high-level language that is very simple to learn. Python is a fairly simple language to code with, and anyone can pick up the fundamentals in a matter of hours or days. Additionally, the language is developer-friendly. 

**Simple to understand.** You will find that learning Python is quite simple. As was already mentioned, Python has a fairly simple syntax. The indentations define the code block instead of semicolons or brackets.

**Language with an Object Focus.** Object-oriented programming is one of Python's primary characteristics. Python offers support for object-oriented languages, classes, object encapsulation, and other related ideas.

**Support for GUI Programming.** With Python, one may create graphical user interfaces by utilizing modules like PyQt5, PyQt4, wxPython, or Tk. The most widely used tool for Python programmers to create graphical applications is PyQt5.

**High-level Language.** Writing programs in Python eliminates the need to memorize memory management and system architecture.

**Portable language.** For instance, we don't need to modify Python code written for Windows in order to run it on Linux, Unix, Mac, or any other platform.

**Large Standard Library.** Because of Python's extensive standard library, which offers a wide range of modules and functions, you don't need to develop your code for everything. Python comes with a lot of libraries, including ones for web browsers, regular expressions, and unit testing.

**Language Typed Dynamically.** Python is a language with dynamic typing. This means that we don't need to define the type of variable because it is selected at run time rather than beforehand (for example, int, double, long, etc.).

You can interact with your notebook using a number of choices available in the Jupyter Notebook. Similar to menus in other programs, the notebook's menu runs down the top. 

- File
- Edit
- View
- Insert
- Cell
- Kernel
- Widgets
- Help

The **File** menu appears first. You can open an existing Notebook or create a new one within it. You can also go here to rename a notebook. The Save and Checkpoint option is the most intriguing menu item. This enables you to set up checkpoints that you may return to in case something goes wrong.

The **Edit** menu comes next. Cells can be copied, cut, and pasted here. In order to remove, divide, or combine a cell, you would also go here. Cells can also be rearranged here. 
 	
You can toggle the header and toolbar visibility using the **View** menu. Additionally, you can turn on or off Line Numbers within cells. If you want to fiddle with the toolbar on the cell, this is the place to go.

The **Insert** menu is just for inserting cells above or below the currently selected cell.

You can execute a single cell, a collection of cells, or every cell using the **Cell** menu. To alter a cell's type, you may also go here, however I think the toolbar is more user-friendly for that. The option to clear a cell's output is another useful feature in this menu. You should clear the output first if you intend to share your Notebook with others so that they can run the cells on their own.

Working with the background-executing kernel is the responsibility of the **Kernel** cell. This is where you can change the kernel that your notebook is running, restart it, reconnect to it, and shut it off. Although it is unlikely that you would work with the kernel frequently, there are occasions when diagnosing a notebook may need restarting the kernel. You would go here if it were to occur.

The widget state can be cleared and saved using the **Widgets** menu. In essence, widgets are JavaScript elements that you can incorporate into your cells to create dynamic content with Python (or an alternative Kernel).

Lastly, there is the **Help** menu, which provides a tour of the Notebook's user interface, a list of keyboard shortcuts, and a wealth of reference information.
